```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/wandb_export_2025-03-04T15_56_54.707-08_00.csv')
df = df[df['test_loss'] < 4e-3]
df

```

```python
variables = ['node_hidden_features', 'node_num_hidden_layers', 'edge_hidden_features', 'edge_num_hidden_layers', 'edge_channels_out', 'learning_rate']

for variable in variables:
    plt.figure(figsize=(10, 6))
    plt.scatter(df[variable], df['test_loss'], label='test_loss')
    plt.xlabel(variable)
    plt.ylabel('test_loss')
    plt.legend()
```
