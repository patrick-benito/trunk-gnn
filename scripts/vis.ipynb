```python
import pandas as pd


df = pd.read_csv("../data/mass_500g_harmonic_ssm/train/raw/data.csv")
```

```python
import numpy as np

# Compute FFT of the 'x30' column
x = df['x30'].values
dt = 0.01  # sampling interval in seconds
n = len(x)
freq = np.fft.rfftfreq(n, d=dt)
fft_vals = np.fft.rfft(x)

# To get amplitude spectrum:
amplitude = np.abs(fft_vals)

# freq contains the frequencies, amplitude contains the corresponding amplitudes
```

```python
import matplotlib.pyplot as plt
plt.plot(freq, amplitude)
plt.xlim(0,5)
```
